```python
#importing libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Loading in data from csv

df = pd.read_csv('movies.csv')
```

```python
df
```

```python
df.sort_values(by='gross',ascending=False)
```

```python
# Looking at missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing*100))
```

```python
filt = df['budget'].isnull()==True
df[filt]
```

```python
#Converting NaN values in budget and gross to 0

budget_filter = df['budget'].isnull()==True
df.loc[budget_filter,'budget']=0

budget_filter = df['gross'].isnull()==True
df.loc[budget_filter,'gross']=0
```

```python
df
```

```python
df.dtypes
```

```python
#Changing data type of budget and gross from float to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
df['country'].unique()
```

```python
filt = df['country']=='India'
df[filt]
```

```python
#Budget vs Gross Correlation

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot Budget vs Gross using seaboarn

sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
df.corr(method='pearson')
```

```python
#Viz - Correlation between different movie features
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Changing string data to numeric categories and using to find correlation

df_numerized = df
df_numerized
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtypes)=='object':
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
```

```python
df_numerized
```

```python
df_numerized.corr()
```

```python
sns.heatmap(df_numerized.corr(),annot=True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
pd.set_option('display.max_rows',None)
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs>0.5]
high_corr
```
